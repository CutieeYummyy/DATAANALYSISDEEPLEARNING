```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#1
data={"Name": pd.Series(["Alice","Bob","Charlie","David","Eva","Frank","Grace","Hannah","Ivan","Jack","Kelly","Liam","Mona","Nina","Oscar"]),
      "Age": pd.Series([25,30,35,28,22,45,34,31,27,29,33,40,26,32,36]),
      "Salary": pd.Series([50000,60000,70000,55000,52000,80000,72000,68000,61000,59000,63000,77000,53000,66000,75000])}
df=pd.DataFrame(data)

print(df)
```

```python
#2
print(df.describe())
```

```python
#3
print(df.loc[df['Age']>28])
```

```python
#4
print(df.aggregate({"Salary": np.average, "Salary": np.mean}))
```

```python
#5
print(df.groupby('Age')['Salary'].mean().sum())
```

```python
#6
print(df.sort_values(by='Salary',ascending=False))
```

```python
#7
age = df['Age'].value_counts()
relative_gender = df['Age'].value_counts(normalize=True) * 100
plt.figure(figsize=(10, 5))
age.plot(kind='bar', color='skyblue')
plt.title('Biểu đồ tần số của biến GT')
plt.xlabel('Giá trị')
plt.ylabel('Tần số')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#8
df['Salary'].plot(marker='D', color='orange', linestyle='-', linewidth=2, markersize=10)
plt.title('Biểu đồ đường cho cột Salary')
plt.xlabel('')
plt.ylabel('')
plt.show()
```

```python
#9
plt.figure(figsize=(6, 6))
df["Age"].plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['blue', 'lightblue'])
plt.title('')
plt.ylabel('')
plt.show()
```

```python
#10
plt.scatter(df['Age'], df['Salary'])
plt.title('Biểu đồ Scatter ')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

```python
#11
missing_info = df.isnull().mean() * 100
print(missing_info)
```

```python
#12

average_age = df[df['Age'] > 30]['Age'].mean()
df.loc[df['Age'] > 30, 'Age'] = average_age

print(df)
```

```python
#13
min_age = df['Age'].min()
max_age = df['Age'].max()

df['Age_normalized'] = (df['Age'] - min_age) / (max_age - min_age)
print(df)
```

```python
#14
df['Age_group'] = pd.cut(df['Age'], bins = [0, 30, 50, float('inf')],labels = ['Young', 'Middle_aged', 'Old'], right=False)
print(df)
```

```python
#15
df['Salary_change'] = df['Salary'] - df['Salary'].iloc[0]
df['Salary_change_percent'] = ((df['Salary'] - df['Salary'].iloc[0]) / df['Salary'].iloc[0]) * 100

print(df)
```

```python
#16
duplicate_rows = df[df.duplicated(subset=['Name'], keep='first')]
df_cleaned = df.drop_duplicates(subset=['Name'], keep='first')

print(df_cleaned)
```
